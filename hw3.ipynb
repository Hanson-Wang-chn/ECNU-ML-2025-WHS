## 第一题

补全以下代码，找出列表中所有大于 10 或为偶数的数，并将它们累加到变量 sum 中，最后输出该变量的值。

```python
numbers = [5, 12, 8, 15, 20, 3]
sum_over_10 = 0
# 请在此处补全代码

```

````python
# 第一题代码
numbers = [5, 12, 8, 15, 20, 3]
sum_over_10 = 0

for num in numbers:
    if num > 10 or num % 2 == 0:
        sum_over_10 += num
sum = sum_over_10
			
print(sum)
````

# 第五题

**Pandas习题**

1. 创建DataFrame  
   创建一个包含以下数据的DataFrame：  

   | Name | Age | City |
   |------|-----|--------|
   | Alice | 25 | New York |
   | Bob | 30 | Los Angeles |
   | Charlie | 35 | Chicago |

2. 数据索引与选择  
   对上述DataFrame df，选择Age列。  
   对上述DataFrame df，选择Age大于30的行。  

3. 列操作  
   对上述DataFrame df，添加一列Salary，值为[50000, 60000, 70000]。  
   对上述DataFrame df，删除City列。  

4. 聚合运算  
   对上述DataFrame df，计算Age列的平均值。  
   对上述DataFrame df，计算Age列的最大值和最小值。

````python
# 第五题（1）代码
df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "City": ["New York", "Los Angeles", "Chicago"]
})
print(df)
````

````python
# 第五题（2）代码
print(df["Age"])
print("\n")
print(df[df["Age"] > 30])
````

````python
# 第五题（3）代码
df["Salary"] = [50000, 60000, 70000]
print(df)
print("\n")

df = df.drop("City", axis=1)
print(df)
````

````python
# 第五题（4）代码
print(df["Age"].mean())  # 30.0
print("\n")
print(df["Age"].max(), df["Age"].min())  # 35 25
````
